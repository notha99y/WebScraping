```python
from bs4 import BeautifulSoup
from requests import get
```

```python
url = 'https://www.srx.com.sg/mrt-home-prices'

res = get(url, timeout=5)
# soup = BeautifulSoup(res.text, 'html.parser')
soup = BeautifulSoup(res.text, 'lxml')
```

```python
# title
soup.title.text
```

```python
# first div tag
soup.div
```

```python
# div tage with attritube id=mrtGraph
stations = soup.find('div', id='mrtGraph').div.find_all('div')
for station in stations:
#     print(station.text)
    station_name = station['class'][0]
    psf = station.text
    print("Station name: {}, PSF: {}".format(station_name, psf))
```
